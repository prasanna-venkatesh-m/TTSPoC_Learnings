```python
import torch
from TTS.utils.manage import ModelManager
from TTS.utils.synthesizer import Synthesizer

# 1. Define your file paths
# Make sure these filenames match exactly what you downloaded
model_path = "/content/tamil.pt"
encoder_path = "/content/speaker_encoder.pt"
vocoder_path = "/content/hifigan.pt"

# AIKosh models usually need a config.json.
# If you don't have one, this script might error out.
config_path = "config.json"

# 2. Initialize the Synthesizer
syn = Synthesizer(
    tts_checkpoint=model_path,
    tts_config_path=config_path,
    speaker_encoder_checkpoint=encoder_path,
    vocoder_checkpoint=vocoder_path,
    use_cuda=torch.cuda.is_available() # Uses GPU if you have one
)

# 3. Your Tamil Text
text = "வணக்கம், இது செயற்கை நுண்ணறிவு மூலம் உருவாக்கப்பட்ட குரல்."

# 4. Generate and Save
print("Synthesizing audio...")
outputs = syn.tts(text)
syn.save_wav(outputs, "tamil_output.wav")
print("Done! Check tamil_output.wav")
```
