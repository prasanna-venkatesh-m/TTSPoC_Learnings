```python
from groq import Groq
from gtts import gTTS
from IPython.display import Audio

# Setup Groq client
client = Groq(api_key="Groq_API_key")

def get_llm_response(user_text):
    response = client.chat.completions.create(
        model="openai/gpt-oss-120b",  # fast & powerful
        messages=[
            {
                "role": "system",
                "content": "You are a helpful assistant.Respond in natural spoken Tamil (colloquial Tamil, like how people speak in Chennai). Avoid formal or literary Tamil. Keep it friendly and simple."
            },
            {
                "role": "user",
                "content": user_text
            }
        ]

    )
    return response.choices[0].message.content


def text_to_speech(text):
    tts = gTTS(text=text, lang='ta')
    tts.save("output.mp3")
    return "output.mp3"


# Run
user_input = input("Enter text: ")

llm_text = get_llm_response(user_input)
print("LLM:", llm_text)

audio_file = text_to_speech(llm_text)

Audio(audio_file)
```
